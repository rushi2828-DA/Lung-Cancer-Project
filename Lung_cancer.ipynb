```python
#corrleation heatmap
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('correlation_heatmap')
plt.show()
```

```python
# Histrogram for all numeric columns

df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Distribtuation of age
sns.distplot(df['AGE'])
plt.show()
```

```python
# Box plot
sns.boxplot(x='LUNG_CANCER', y='AGE', data=df)
plt.title('Distribution of Age by Lung Cancer Status')
plt.xlabel('Lung Cancer')
plt.ylabel('Age')
plt.show()
```

```python
# histrogram of lung cancer vs age

# Assuming 'AGE' is a numerical column and 'LUNG_CANCER' is a categorical column (e.g., 0 for no, 1 for yes)
# Create a histogram of age, colored by the lung cancer status
sns.histplot(data=df, x='AGE', hue='LUNG_CANCER', multiple='stack')
plt.title('Histogram of Age by Lung Cancer Status')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# relationship between age and lung cancer

plt.figure(figsize=(8, 6))
sns.barplot(x='LUNG_CANCER', y='AGE', data=df)
plt.title('Relationship between Age and Lung Cancer')
plt.xlabel('Lung Cancer')
plt.ylabel('Age')
plt.show()
```

```python
# spending by gender

plt.figure(figsize=(8, 6))
sns.countplot(x='GENDER', data=df)
plt.title('Spending by Gender')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```
